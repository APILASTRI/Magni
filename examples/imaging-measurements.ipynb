# `imaging-measurements`: Show examples of how to use the available sampling patterns

This example primarily shows the usage of the following functions:
- `magni.imaging.measurements.lissajous_sample_image`
- `magni.imaging.measurements.lissajous_sample_surface`
- `magni.imaging.measurements.plot_pattern`
- `magni.imaging.measurements.plot_pixel_mask`
- `magni.imaging.measurements.random_line_sample_image`
- `magni.imaging.measurements.random_line_sample_surface`
- `magni.imaging.measurements.spiral_sample_image`
- `magni.imaging.measurements.spiral_sample_surface`
- `magni.imaging.measurements.square_spiral_sample_image`
- `magni.imaging.measurements.square_spiral_sample_surface`
- `magni.imaging.measurements.uniform_line_sample_image`
- `magni.imaging.measurements.uniform_line_sample_surface`
- `magni.imaging.measurements.uniform_rotated_line_sample_image`
- `magni.imaging.measurements.uniform_rotated_line_sample_surface`
- `magni.imaging.measurements.unique_pixels`
- `magni.imaging.measurements.zigzag_sample_image`
- `magni.imaging.measurements.zigzag_sample_surface`

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

import magni
```

Sample the surface of a physical sample (random lines)

```python
l = 1e-6; w = 1.5e-6; speed = 4.3e-7; time = 27.0; sample_rate = 3.0; discrete=None; seed=6021;
surface_coords = magni.imaging.measurements.random_line_sample_surface(l, w, speed, sample_rate, time, discrete=discrete, seed=seed)
magni.utils.plotting.setup_matplotlib({'figure': {'figsize': (20, 10)}}, cmap='gray')
magni.imaging.measurements.plot_pattern(l, w, surface_coords, 'surface')

# Set transparent background on image
plt.gcf().patch.set_alpha(0.0)
plt.gca().patch.set_alpha(0.0)
```

Sample an image (spiral)

```python
h = 256; w = 256; scan_length = 3000.0; num_points = 4000;
image_coords = magni.imaging.measurements.spiral_sample_image(h, w, scan_length, num_points)
magni.imaging.measurements.plot_pattern(h, w, image_coords, 'image')

# Set transparent background on image
plt.gcf().patch.set_alpha(0.0)
plt.gca().patch.set_alpha(0.0)
```

Plot the pixel mask of the spiral above

```python
unique_pixels = magni.imaging.measurements.unique_pixels(image_coords)
magni.imaging.measurements.plot_pixel_mask(h, w, unique_pixels)

# Set transparent background on image
plt.gcf().patch.set_alpha(0.0)
plt.gca().patch.set_alpha(0.0)
```

Plot various other patterns

```python
l = 1e-6; w = 1e-6; speed = 4.3e-7; time = 27.0; sample_rate = 10.0;
titles = ['random line', 'uniform line', 'uniform rotated line (follow_edge=True)',
          'uniform rotated line (follow_edge=False)', 'zigzag', 'square spiral', 'spiral (rect_area=False)',
          'spiral (rect_area=True)', 'lissajous']
pattern_coords = [
    magni.imaging.measurements.random_line_sample_surface(l, w, speed, sample_rate, time, discrete=100),
    magni.imaging.measurements.uniform_line_sample_surface(l, w, speed, sample_rate, time),
    magni.imaging.measurements.uniform_rotated_line_sample_surface(l, w, speed, sample_rate, time, angle=2. / 3 * np.pi,
                                                                   follow_edge=True),
    magni.imaging.measurements.uniform_rotated_line_sample_surface(l, w, speed, sample_rate, time, angle=2. / 3 * np.pi,
                                                                   follow_edge=False),
    magni.imaging.measurements.zigzag_sample_surface(l, w, speed, sample_rate, time),
    magni.imaging.measurements.square_spiral_sample_surface(l, w, speed, sample_rate, time),
    magni.imaging.measurements.spiral_sample_surface(l, w, speed, sample_rate, time, rect_area=False),
    magni.imaging.measurements.spiral_sample_surface(l, w, speed, sample_rate, time, rect_area=True),
    magni.imaging.measurements.lissajous_sample_surface(l, w, speed, sample_rate, time, f_x=0.25, f_y=0.35, speed_mode=2)]

for title, coords in zip(titles, pattern_coords):
    magni.imaging.measurements.plot_pattern(l, w, coords, 'surface')
    plt.title(title, y=1.05)

    # Set transparent background on image
    plt.gcf().patch.set_alpha(0.0)
    plt.gca().patch.set_alpha(0.0)
```
