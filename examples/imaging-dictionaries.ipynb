# `imaging-dictionaries`: Show an example of how to use dictionaries

This example shows how to use the fast-ops dictionaries as well as a collection of functions for analysing the potential of a dictionary.

This example primarily shows the usage of the following functions:

- `magni.imaging.dictionaries.get_DCT`
- `magni.imaging.dictionaries.get_DFT`
- `magni.imaging.dictionaries.analysis.show_coefficient_histogram`
- `magni.imaging.dictionaries.analysis.show_transform_coefficients`
- `magni.imaging.dictionaries.analysis.show_sorted_coefficients`
- `magni.imaging.dictionaries.analysis.get_reconstructions`
- `magni.imaging.dictionaries.analysis.show_psnr_energy_rolloff`
- `magni.imaging.dictionaries.analysis.show_reconstructions`
- `magni.imaging.dictionaries.analysis.show_transform_quantiles`

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

import magni
```

```python
# Load image
img = magni.afm.io.read_mi_file('example.mi').get_buffer('Topography')[0].data
img = magni.imaging.visualisation.stretch_image(img, 1)

# Convert image to vector
h, w = img.shape
vec = magni.imaging.mat2vec(img)
```

Transform the image, remove small coefficients, transform the resulting image back again, and compare it to the original

```python
# Make dictionary, transform image, remove small coefficients, and transform back again
Psi = magni.imaging.dictionaries.get_DCT((h, w))
coefs = Psi.T.dot(vec)
coefs[np.abs(coefs) < 0.05] = 0
vec = Psi.dot(coefs)

print('\n')
print('{:6,} DCT coefficients'.format(len(coefs)))
print('{:6,} DCT coefficients with absolute value greater than or equal to 0.05'.format(np.sum(np.abs(coefs) >= 0.05)))
print('\n')
```

```python
# Show the original image
magni.utils.plotting.setup_matplotlib({'figure': {'figsize': (18, 10)}})
fig, axes = plt.subplots(1, 2)
magni.imaging.visualisation.imshow(img, ax=axes[0])
axes[0].set_title('Original image')

# Show the transform coded image
modified_img = magni.imaging.vec2mat(vec, (h, w))
magni.imaging.visualisation.imshow(modified_img, ax=axes[1], intensity_func=magni.imaging.visualisation.stretch_image,
                                   intensity_args=(1,))
_ = axes[1].set_title('Modified image')

# Set transparent background on image
fig.patch.set_alpha(0.0)
axes[0].patch.set_alpha(0.0)
axes[1].patch.set_alpha(0.0)
```

```python
# Show histogram of transform coefficients for both DFT and DCT transforms
transforms=('DFT', 'DCT')
magni.imaging.dictionaries.analysis.show_coefficient_histogram(img, transforms, range=(-6, 6))

# Set transparent background on image
plt.gcf().patch.set_alpha(0.0)
plt.gca().patch.set_alpha(0.0)
```

```python
# Show the transform coefficients
magni.imaging.dictionaries.analysis.show_transform_coefficients(img, transforms)

# Set transparent background on image
fig = plt.gcf()
fig.patch.set_alpha(0.0)
for ax in fig.get_axes():
    ax.patch.set_alpha(0.0)
```

```python
# Show the transform coefficients in sorted order
magni.imaging.dictionaries.analysis.show_sorted_coefficients(img, transforms)

# Set transparent background on image
plt.gcf().patch.set_alpha(0.0)
plt.gca().patch.set_alpha(0.0)
```

```python
# Show PSNR and energy rolloff for reconstructions from a DCT
transform = 'DCT'
fractions = [0.005, 0.01, 0.02, 0.05, 0.1, 0.15, 0.2, 0.5]
coefficients, reconstructions = magni.imaging.dictionaries.analysis.get_reconstructions(img, transform, fractions)
psnr, energy = magni.imaging.dictionaries.analysis.show_psnr_energy_rolloff(img, reconstructions, fractions, return_vals=True)

# Set transparent background on image
fig = plt.gcf()
fig.patch.set_alpha(0.0)
for ax in fig.get_axes():
    ax.patch.set_alpha(0.0)
```

```python
# Show coefficients used in reconstructions and the reconstructions from using a DCT
magni.imaging.dictionaries.analysis.show_reconstructions(coefficients, reconstructions, transform, fractions)

# Set transparent background on image
for fig_num in plt.get_fignums():
    fig = plt.figure(fig_num)
    fig.patch.set_alpha(0.0)
    for ax in fig.get_axes():
        ax.patch.set_alpha(0.0)
```

```python
# Show the 10% largest DCT coefficients and their distribution within the four quantiles. Overlay the figure with a model mask.
fraction = 0.1
x, y = np.meshgrid(*map(np.arange, img.shape))
disc_mask = np.sqrt(x**2 + y**2) <= 100
_ = magni.imaging.dictionaries.analysis.show_transform_quantiles(img, transform, fraction, area_mask=disc_mask)

# Set transparent background on image
fig = plt.gcf()
fig.patch.set_alpha(0.0)
for ax in fig.get_axes():
    ax.patch.set_alpha(0.0)
```

```python
# Make over-complete dictionary, transform image, remove small coefficients, and transform back again
vec = magni.imaging.mat2vec(img)  # Reset original image
Psi_oc = magni.imaging.dictionaries.get_DFT((h, w), (2*h, 2*w))
coefs_oc = Psi_oc.conj().T.dot(vec)
coefs_oc[np.abs(coefs_oc) < 0.05] = 0
vec = Psi_oc.dot(coefs_oc)

print('\n')
print('{:6,} DFT coefficients'.format(len(coefs_oc)))
print('{:6,} DFT coefficients with absolute value greater than or equal to 0.05'.format(np.sum(np.abs(coefs_oc) >= 0.05)))
print('\n')
```

```python
# Show the original image
magni.utils.plotting.setup_matplotlib({'figure': {'figsize': (18, 10)}})
fig, axes = plt.subplots(1, 2)
magni.imaging.visualisation.imshow(img, ax=axes[0])
axes[0].set_title('Original image')

# Show the transform coded image
modified_img = magni.imaging.vec2mat(vec, (h, w))
magni.imaging.visualisation.imshow(modified_img, ax=axes[1], intensity_func=magni.imaging.visualisation.stretch_image,
                                   intensity_args=(1,))
_ = axes[1].set_title('Modified image')

# Set transparent background on image
fig.patch.set_alpha(0.0)
axes[0].patch.set_alpha(0.0)
axes[1].patch.set_alpha(0.0)
```
