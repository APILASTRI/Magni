# `magni`: Show an example of multiple parts of the package put to use

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

import display
import magni
```

```python
# Load image
img = magni.afm.io.read_mi_file('example.mi').get_buffer('Topography')[0].data
img = magni.imaging.visualisation.stretch_image(img, 1)
img_org = img

# Convert image to vector
h, w = img.shape
vec = magni.imaging.mat2vec(img)
vec -= vec.mean()

# Show image
magni.utils.plotting.setup_matplotlib(settings={'figure': {'figsize': (9, 9)}})
magni.imaging.visualisation.imshow(img_org)
_ = plt.title('Original image')

# Set transparent background on image
plt.gcf().patch.set_alpha(0.0)
plt.gca().patch.set_alpha(0.0)
```

Select spiral sampling pattern and DCT dictionary, and instantiate a MultiDomainImage

```python
# Choose sampling pattern handles, args, and kwargs
handles = [
    magni.imaging.measurements.uniform_rotated_line_sample_image,
    magni.imaging.measurements.lissajous_sample_image,
    magni.imaging.measurements.spiral_sample_image]

scan_length = 0.15 * 2 * h * w
num_points = 10 * int(scan_length)
args = (h, w, scan_length, num_points)

kwargss = [
    {'angle': 2.574812},
    {'f_y': 10.365261, 'f_x': 9.951238, 'theta_y': 1.100845, 'theta_x': 1.603954},
    {'rect_area': True}]

Phis = []

# Choose DCT dictionary
Psi = magni.imaging.dictionaries.get_DCT((h, w))

# Use Domain class to move between domains
domains = []

# Show original image
magni.utils.plotting.setup_matplotlib(settings={'figure': {'figsize': (18, 18)}})
fig, axes = plt.subplots(2, 2)
magni.imaging.visualisation.imshow(img_org, ax=axes[0, 0])
axes[0, 0].set_title('Original image')

titles = ['Uniform rotated line', 'Lissajous', 'Spiral']

# Store characteristics for later comparison
mse_vals = []
psnr_vals = []

for handle, kwargs, ax, title in zip(handles, kwargss, axes.ravel()[1:], titles):
    # Construct sampling pattern
    img_coords = handle(*args, **kwargs)
    Phi = magni.imaging.measurements.construct_measurement_matrix(img_coords, h, w)
    Phis.append(Phi)

    # Construct domain
    domain = magni.imaging.domains.MultiDomainImage(Phi, Psi)
    domain.measurements = Phi.dot(vec)
    domains.append(domain)

    # Show sampled image
    img = magni.imaging.vec2mat(domain.image, (h, w))
    magni.imaging.visualisation.imshow(img, ax=ax, intensity_func=magni.imaging.visualisation.stretch_image,
                                       intensity_args=(1.0,))
    ax.set_title(title + ' sampled image')

    # Compute characteristics
    mse_vals.append(magni.imaging.evaluation.calculate_mse(vec, domain.image))
    psnr_vals.append(magni.imaging.evaluation.calculate_psnr(vec, domain.image, float(np.max(np.abs(vec)))))

    # Set transparent background on image
    ax.patch.set_alpha(0.0)

# Set transparent background on image
fig.patch.set_alpha(0.0)
```

Reconstruct the image from the spiral sampled measurements

```python
# Setup weighted soft thresholding
X = np.row_stack([np.repeat(np.linspace(0, 1, w), h), np.tile(np.linspace(0, 1, h), w)])
W = 0.0015 * np.exp(-((X - 0.017) * np.float_([[6.6, 1.3], [1.3, 6.6]]).dot(X - 0.017)).sum(axis=0))
magni.cs.reconstruction.it.config.update({
    'kappa_fixed': 0.6,
    'threshold': 'fixed',
    'threshold_fixed': int(0.05 * h * w),
    'threshold_operator': 'weighted_soft',
    'threshold_weights': W.reshape((-1, 1))
})
print('Reconstruction algorithm used: {}'.format(magni.afm.config['algorithm']))

# Show original image
fig, axes = plt.subplots(2, 2)
magni.imaging.visualisation.imshow(img_org, ax=axes[0, 0])
axes[0, 0].set_title('Original image')

for domain, Phi, ax, title in zip(domains, Phis, axes.ravel()[1:], titles):
    # Reconstruct
    domain.image = magni.afm.reconstruction.reconstruct(domain.measurements, Phi, Psi)

    # Show reconstructed image
    img = magni.imaging.vec2mat(domain.image, (h, w))
    magni.imaging.visualisation.imshow(img, ax=ax, intensity_func=magni.imaging.visualisation.stretch_image,
                                       intensity_args=(1.0,))
    ax.set_title('Reconstructed ' + title.lower() + ' sampled image')

    # Set transparent background on image
    ax.patch.set_alpha(0.0)

# Set transparent background on image
fig.patch.set_alpha(0.0)
```

Calculate metrics of the reconstruction quality

```python
for Phi in Phis:
    # Analyse
    mse, psnr = magni.afm.reconstruction.analyse(vec, Phi, Psi)
    mse_vals.append(mse)
    psnr_vals.append(psnr)

mse_val_strs = list(map('{:.2e}'.format, mse_vals))
psnr_val_strs = list(map('{:.2f}'.format, psnr_vals))
display.print_table((titles, mse_val_strs[:3], psnr_val_strs[:3], mse_val_strs[3:], psnr_val_strs[3:]),
                    ('Sampling pattern', 'MSE, sampled', 'PSNR, sampled', 'MSE, reconstructed', 'PSNR, reconstructed'),
                    ('|', '', '|', ''))
```
