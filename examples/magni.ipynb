```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

import display
import magni
```

```python
# Load image
img = magni.afm.io.read_mi_file('example.mi').get_buffer('Topography')[0].get_data(magni.imaging.visualisation.stretch_image, (1,))

# Convert image to vector
h, w = img.shape
vec = magni.imaging.mat2vec(img)
vec -= vec.mean()

# Show image
magni.utils.plotting.setup_matplotlib(settings={'figure': {'figsize': (8, 8)}}, cmap='gray')
magni.imaging.visualisation.imshow(img)
_ = plt.title('Original image')
```

```python
# Choose spiral sampling pattern
scan_length = 0.1 * 2 * h * w
num_points = 10 * int(scan_length)  # Make sure to have enough points
img_coords = magni.imaging.measurements.spiral_sample_image(h, w, scan_length, num_points)
Phi = magni.imaging.measurements.construct_measurement_matrix(img_coords, h, w)

# Choose DCT dictionary
Psi = magni.imaging.dictionaries.get_DCT((h, w))

# Use Domain class to move between domains
domain = magni.imaging.domains.MultiDomainImage(Phi, Psi)
domain.measurements = Phi.dot(vec)

# Show sampled image
img = magni.imaging.vec2mat(domain.image, (h, w))
magni.imaging.visualisation.imshow(img, intensity_func=magni.imaging.visualisation.stretch_image, intensity_args=(1.0,))
plt.title('Sampled image')

# Store characteristics for later comparison
mse_vals = [magni.imaging.evaluation.calculate_mse(vec, domain.image)]
psnr_vals = [magni.imaging.evaluation.calculate_psnr(vec, domain.image, float(np.max(np.abs(vec))))]
```

```python
# Reconstruct
domain.image = magni.afm.reconstruction.reconstruct(domain.measurements, Phi, Psi)
print('Reconstruction algorithm used: {}'.format(magni.afm.config.get('algorithm')))

# Show reconstructed image
img = magni.imaging.vec2mat(domain.image, (h, w))
magni.imaging.visualisation.imshow(img, intensity_func=magni.imaging.visualisation.stretch_image, intensity_args=(1.0,))
_ = plt.title('Reconstructed image')
```

```python
# Analyse
mse, psnr = magni.afm.reconstruction.analyse(vec, Phi, Psi)
mse_vals = ['{:.2e}'.format(val) for val in mse_vals + [mse]]
psnr_vals = ['{:.2f}'.format(val) for val in psnr_vals + [psnr]]

display.print_table((('Sampled', 'Reconstructed'), mse_vals, psnr_vals), ('Image', 'MSE', 'PSNR'), ('|', ''))
```
