```python
%matplotlib inline

import matplotlib.pyplot as plt

import magni
```

```python
# Load example data
img = magni.afm.io.read_mi_file('example.mi').get_buffer('Topography')[0].get_data(magni.imaging.visualisation.stretch_image, (1,))

# Setup matplotlib using magni with custom figure size
magni.utils.plotting.setup_matplotlib({'figure': {'figsize': (16, 8)}})

# Plot using default colour maps
fig, axes = plt.subplots(1, 2)
magni.imaging.visualisation.imshow(img, ax=axes[0])
axes[0].set_title('Default colour map')

magni.imaging.visualisation.imshow(img, ax=axes[1], cmap='gray')
_ = axes[1].set_title('Gray colour map')
```

```python
# Plot using seq colour maps
magni.utils.plotting.setup_matplotlib({'figure': {'figsize': (12, 36)}})
fig, axes = plt.subplots(int(len(magni.utils.plotting.seq_cmaps)/2), 2)

for k in range(int(len(magni.utils.plotting.seq_cmaps)/2)):
    cmap = magni.utils.plotting.seq_cmaps[2*k]
    magni.imaging.visualisation.imshow(img, ax=axes[k][0], cmap=cmap)
    axes[k][0].set_title('{0} (seq_cmaps[{1}]'.format(cmap, 2*k))
    cmap = magni.utils.plotting.seq_cmaps[2*k+1]
    magni.imaging.visualisation.imshow(img, ax=axes[k][1], cmap=cmap)
    axes[k][1].set_title('{0} (seq_cmaps[{1}])'.format(cmap, 2*k+1))
```

```python
# Plot using div colour maps
magni.utils.plotting.setup_matplotlib({'figure': {'figsize': (12, 32)}})
fig, axes = plt.subplots(int(len(magni.utils.plotting.div_cmaps)/2), 2)

for k in range(int(len(magni.utils.plotting.div_cmaps)/2)):
    cmap = magni.utils.plotting.div_cmaps[2*k]
    magni.imaging.visualisation.imshow(img, ax=axes[k][0], cmap=cmap)
    axes[k][0].set_title('{0} (div_cmaps[{1}])'.format(cmap, 2*k))
    cmap = magni.utils.plotting.div_cmaps[2*k+1]
    magni.imaging.visualisation.imshow(img, ax=axes[k][1], cmap=cmap)
    axes[k][1].set_title('{0} (div_cmaps[{1}])'.format(cmap, 2*k+1))
```
