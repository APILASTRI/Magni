```python
# Save phase transition plot as pdf

iht_pt = magni.cs.phase_transition.io.load_phase_transition('data.hdf5', 'rad_iht')
sl0_mod_pt = magni.cs.phase_transition.io.load_phase_transition('data.hdf5', 'rad_sl0_mod')

curves = [{'delta': iht_pt[0], 'rho': iht_pt[1], 'label': 'IHT'},
          {'delta': sl0_mod_pt[0], 'rho': sl0_mod_pt[1], 'label': 'SL0 MOD'}]

_ = magni.cs.phase_transition.plotting.plot_phase_transitions(curves, plot_l1=True) # 'phase_transitions.pdf'
```

```python
# Save phase transition colormaps as pdf
magni.utils.plotting.setup_matplotlib(cmap=magni.utils.plotting.div_cmaps[5])

with tables.open_file('data.hdf5', mode='r') as h5file:
    iht_dist = h5file.get_node('/rad_iht/dist').read()
    sl0_mod_dist = h5file.get_node('/rad_sl0_mod/dist').read()

delta = np.linspace(0, 1, iht_dist.shape[0] + 1)[1:]
rho = np.linspace(0, 1, iht_dist.shape[1] + 1)[1:]

magni.cs.phase_transition.plotting.plot_phase_transition_colormap(iht_dist, delta, rho) # 'iht_pt_colormap.pdf'
magni.cs.phase_transition.plotting.plot_phase_transition_colormap(sl0_mod_dist, delta, rho) # 'sl0_mod_pt_colormap.pdf'
```
